```python
from mtcnn import MTCNN
from PIL import Image
from os import listdir
from os.path import isdir
from numpy import asarray

detector = MTCNN()

def flip_image(image):
    
    img = image.transpose(Image.FLIP_LEFT_RIGHT)
    
    return img

def extrair_face(arquivo, size = (160,160)):
    
    img = Image.open(arquivo)
    img = img.convert('RGB')

    array = asarray(img)

    results = detector.detect_faces(array)

    x1, y1, width, height = results[0]['box']
    x2, y2 = x1 + width, y1 + height

    face = array[y1:y2, x1:x2]

    image = Image.fromarray(face)
    image = image.resize(size)

    return image

def load_fotos(directory_scr, directory_target):
    
    for filename in listdir(directory_scr):
        
        path = directory_scr + filename
        path_tg = directory_target + filename
        path_tg_flip = directory_target + 'flip_'+filename

        try:
            face = extrair_face(path)
#             print("face foi extraida")
            face_flip = flip_image(face)
#             print("flipando imagem")
            face_flip.save(path_tg_flip, "JPEG", quality = 100, optimize = True, progressive = True)
#             print("salvando imagem flipada")
            face.save(path_tg, "JPEG", quality = 100, optimize = True, progressive = True)
#             print(f"imagem salva")

        except:
            print(f"\nerror na imagem {path}")

def load_dir(directory_scr, directory_target):
    
    for subdir in listdir(directory_scr):

        path = directory_scr + "/" + subdir + "/"
        path_tg = directory_target + "/" + subdir + "/"

        if not isdir(path):
            continue
    
        load_fotos(path, path_tg)


if __name__ == "__main__":
    print('Buscando fotos no path de origem')
#     load_dir("/home/vitor/Documents/PROGRAMAS/Programas_Python/Condominio_mineiro/full_images",
#              "/home/vitor/Documents/PROGRAMAS/Programas_Python/Condominio_mineiro/faces")
    load_dir("/home/vitor/Documents/PROGRAMAS/Programas_Python/Condominio_mineiro/full_images_validation",
             "/home/vitor/Documents/PROGRAMAS/Programas_Python/Condominio_mineiro/faces_validation")
    
/home/vitor/Documents/PROGRAMAS/Programas_Python/Condominio_mineiro/full_images_validation/steferson
```
