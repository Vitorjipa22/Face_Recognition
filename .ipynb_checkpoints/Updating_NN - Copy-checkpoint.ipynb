```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import collections

from sklearn.utils import shuffle
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

from tensorflow.keras.utils import to_categorical
from tensorflow.keras import models
from tensorflow.keras import layers
from tensorflow.keras import Sequential

from sklearn.preprocessing import Normalizer

%matplotlib inline
```

```python
def print_confusion_matrix(model_name, valY, yhat_val):
    
    cm = confusion_matrix(valY, yhat_val)
    fig, ax = plot_confusion_matrix(conf_mat = cm , figsize=(5,5))
    plt.show()
```

```python
df = pd.read_csv("faces.csv")
df = df.drop(["Unnamed: 0"], axis = 1)
df.head()
```

```python
df_novo = pd.read_csv("novo_morador.csv")
df_novo = df_novo.drop(["Unnamed: 0"], axis = 1)
df_novo.head()
```

```python
df = pd.concat([df,df_novo])
df.head()
```

```python
X = np.array(df.drop(["target"], axis=1))
X.shape
```

```python
y = np.array(df.target)
y.shape
print(y[0])
```

```python
trainX, trainY = shuffle(X, y, random_state = 0) # misturando tudo 
a = 0
for i in range(len(trainY)):
    print(i) if trainY[i]  != "desconhecidos" else 0
    
#print(collections.Counter(trainY))
```

## TRATANDO AS LABELS

```python
out_encoder = LabelEncoder() # enumerando as saidas
out_encoder.fit(trainY)
```

```python
trainY = out_encoder.transform(trainY)
n_classes = len(np.unique(trainY))
n_classes
```

### KERAS

```python
trainY = to_categorical(trainY)
trainY
```

```python
model = models.Sequential()
model.add(layers.Dense(64, activation = 'relu', input_shape = (128,)))
model.add(layers.Dropout(0.5))
model.add(layers.Dense(n_classes, activation = 'softmax'))
 
model.summary()
```

```python
model.compile(optimizer="adam", loss= "categorical_crossentropy", metrics=['accuracy'])
```

```python
model.fit(trainX, trainY, epochs=20, batch_size = 8)
```

```python
df.to_csv('faces.csv')
model.save("faces.h5")
model.save_weights('facesWeights.h5')
```
